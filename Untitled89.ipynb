```python
def checkoutlier(column_name):
    q1=df[column_name].quantile(.25)
    q3=df[column_name].quantile(.75)
    iqr=q3-q1
    lowerbound=q1-1.5*iqr
    upperbound=q3+1.5*iqr
    outlier_mask=((df[column_name] < lowerbound) | (df[column_name] > upperbound ))
    return df[outlier_mask]
```

```python
checkoutlier('TotalCharges')
```

```python
checkoutlier('MonthlyCharges')
```

```python
checkoutlier('tenure')
```

```python
def checkoutlier2(column_name):
    mean=df[column_name].mean()
    std=df[column_name].std()
    z_score=(df[column_name]-mean)/std
    z_threshold=3
    return df[z_score.abs()>z_threshold]
```

```python
checkoutlier2('tenure')
```
